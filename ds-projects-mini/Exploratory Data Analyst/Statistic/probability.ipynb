## **Soal 1 a**

```python
pmf = df['NumWebVisitsMonth'].value_counts(normalize=True).sort_index() * 100
pmf
```

## **Soal 1 b**

```python
# Probability of having exactly 5 web visits
prob_5_visits = pmf.get(5, 0)

# Probability of having more than 5 web visits
prob_more_than_5_visits = pmf[pmf.index > 5].sum()

prob_5_visits, prob_more_than_5_visits
```

**Probabilitas seorang pelanggan memiliki 5 kunjungan web bulanan adalah sekitar 12.54%. Sementara itu, probabilitas seorang pelanggan memiliki lebih dari 5 kunjungan web bulanan adalah sekitar 52.23%. Jadi, pelanggan lebih mungkin memiliki lebih dari 5 kunjungan daripada tepat 5 kunjungan.**

## **Soal 2**

```python
import pandas as pd
import scipy.stats as stats


# Drop rows with missing income or education values
df = df.dropna(subset=['Income', 'Education'])

# Define the two education groups to compare, e.g., "Graduation" vs "PhD"
group1 = df[df['Education'] == 'Graduation']['Income']
group2 = df[df['Education'] == 'PhD']['Income']

# Perform the two-sample t-test
t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=False)

# Display the results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Terdapat perbedaan signifikan dalam rata-rata pendapatan antara dua kelompok pendidikan ini.")
else:
    print("Tidak terdapat perbedaan signifikan dalam rata-rata pendapatan antara dua kelompok pendidikan ini.")
```

**Dapat disimpulkan bahwa tingkat pendidikan ("Graduation" vs. "PhD") berpengaruh signifikan terhadap pendapatan. Ini mungkin menunjukkan bahwa tingkat pendidikan yang lebih tinggi (seperti PhD) terkait dengan pendapatan yang lebih tinggi atau berbeda secara signifikan dibandingkan dengan tingkat pendidikan yang lebih rendah (seperti Graduation).**
